# Creating a Linear Regression Model using Statsmodels and Sklearn¶

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
y = df['Happiness Score']
x = df[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom',
      'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 1, test_size = 0.2)
```

```python
x_const=sm.add_constant(x)
model=sm.OLS(y,x_const).fit()
```

```python
model.summary()
```

# Model Building

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
print(f'R2 score for train: {lr.score(x_train, y_train)}')
print(f'R2 score for test: {lr.score(x_test, y_test)}')
```

```python
y_pred_test = lr.predict(x_test)
y_pred_train = lr.predict(x_train)
```

```python
print(f'mean squared error train : {mean_squared_error(y_train, y_pred_train)}')
print(f'mean squared error test : {mean_squared_error(y_test, y_pred_test)}')
```
